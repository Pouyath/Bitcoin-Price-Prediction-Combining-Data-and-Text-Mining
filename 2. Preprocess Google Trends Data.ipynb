```python
import pandas as pd
```

```python
list_ = []
for i in range(1,15):
    try:
        df = pd.read_csv('Data/google_trends/google_trends'+str(i)+'.csv', skiprows=[0])
        print (len(df))
        list_.append(df)
    except:
        print ('Problem with :',i)

df = pd.concat(list_)
df = df.reset_index()
# df.dropna(how='any', inplace=True)
print (len(df))
```

```python
df = df.drop_duplicates()
print(len(df))
```

```python
google_trends = pd.DataFrame()
google_trends['Timestamp'] = df['Ημέρα']
google_trends['Trend_Values'] = df['Bitcoin: (Παγκοσμίως)']
```

```python
google_trends.to_csv('google_trends_raw_day.csv')
len(google_trends)
```
